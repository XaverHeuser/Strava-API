```python
from utils.api_data import get_data_activities, get_activity_by_id
```

# Get all activities and their ids

```python
request_page_num = 1
all_activities = []

while True:
    param = {'per_page': 200, 'page': request_page_num} 
    my_dataset = get_data_activities(param)
    if len(my_dataset) == 0:
        break

    if all_activities:
        all_activities.extend(my_dataset)

    else:
        all_activities = my_dataset

    request_page_num += 1

# print name, start_date, distance, type, id
for activity in all_activities:
    print(activity['name'], activity['start_date'], activity['distance'], activity['type'], activity['id'])
```
